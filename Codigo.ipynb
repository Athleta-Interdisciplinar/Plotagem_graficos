# Plotando gráficos

```python
# Importando bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Mostrando o df
df = pd.read_csv('Athletes_summer_games.csv')
display(df)
```

```python
# Informações básicas do DataFrame
print(df.info())
```

### Gráficos

#### 1 - Atletas por Sexo 

```python
# Contagem de atletas por sexo
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sex', palette='Set2')
plt.title('Distribuição de Sexo dos Atletas')
plt.show()
```

Neste gráfico, observamos uma quantidade significativamente maior de homens em comparação com as mulheres. É possível que existam regiões onde o esporte masculino é mais promovido ou onde a participação masculina é maior. Identificar essas diferenças regionais pode ajudar a compreender melhor os fatores culturais e econômicos que influenciam a participação no esporte. Considere também as oportunidades disponíveis para cada gênero. Há barreiras estruturais ou culturais que possam limitar a participação das mulheres? Existe uma oferta adequada de programas e apoio para as mulheres no esporte? Há políticas ou iniciativas que possam estar influenciando a disparidade na participação entre os sexos?

#### 2 - Distribuição de idade

```python
# Distribuição de Idade
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribuição de Idades dos Atletas')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

Neste gráfico, observamos que a maior parte dos atletas está na faixa etária de 20 a 30 anos. Esse padrão pode refletir a etapa da vida em que muitos atletas estão em sua forma física ideal e têm a energia e a capacidade de treinamento necessárias para competir em alto nível. Além disso, atletas jovens podem estar no início de suas carreiras esportivas, enquanto aqueles na faixa de 30 a 40 anos podem estar no auge de suas habilidades ou se aproximando da aposentadoria. É importante considerar o impacto das diferentes fases da vida no envolvimento esportivo. Atletas mais jovens podem estar em busca de oportunidades para ganhar experiência e se estabelecer no esporte, enquanto atletas mais velhos podem estar consolidando suas carreiras ou se dedicando a esportes menos exigentes fisicamente.

#### 3 - Número de participações ao longo doa anos

```python
# Número de participações ao longo doa anos
contador_anos = df['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=contador_anos.index, y=contador_anos.values, marker='o',color='Orange')
plt.title('Número de Participações ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Número de Participações')
plt.show()
```

Neste gráfico, observamos uma tendência clara de crescimento no número de participações ao longo dos anos. Esse aumento contínuo pode refletir vários fatores positivos, como o crescimento da popularidade dos eventos esportivos, a expansão das oportunidades para atletas e o aumento geral do interesse em competições esportivas.

Além disso, o aumento na participação pode ser um reflexo de campanhas bem-sucedidas para atrair novos talentos e encorajar a participação de diversos grupos demográficos. Pode também indicar um crescente reconhecimento da importância do esporte para a saúde e bem-estar, resultando em um maior envolvimento da comunidade.

#### 4 - Distribução de Esportes

```python
# Distribução de Esportes
contador_esporte = df['Sport'].value_counts().reset_index()
contador_esporte.columns = ['Sport', 'Count']

plt.figure(figsize=(12, 8))
sns.barplot(data= contador_esporte.head(20), x='Count', y='Sport', palette='viridis')
plt.title('Top 20 Esportes com Mais Participações')
plt.show()
```

Neste gráfico, observamos que os esportes mais praticados são Atletismo (Athletics), Ginástica (Gymnastics), Natação (Swimming), Tiro (Shooting) e Esgrima (Fencing). A predominância desses esportes pode refletir vários fatores, como sua popularidade global, tradição e acessibilidade.

O Atletismo, por exemplo, é frequentemente considerado a base de muitos eventos esportivos, oferecendo uma ampla gama de modalidades que atraem uma grande diversidade de atletas. A Ginástica, com suas diferentes disciplinas, também é bastante popular devido à sua combinação de habilidade técnica e apresentação artística. A Natação é amplamente praticada devido ao seu impacto positivo na saúde e sua inclusão em muitos programas de treinamento esportivo.

#### 5 - Análise de Medalhas por Esporte e Sexo

```python
# Análise de Medalhas por Esporte e Sexo
medalha_esporte_sexo = df[df['Medal'].notna()].groupby(['Sport', 'Sex']).size().reset_index(name='Count')

plt.figure(figsize=(14, 8))
sns.catplot(data= medalha_esporte_sexo, x='Count', y='Sport', hue='Sex', kind='bar', palette='muted', height=8, aspect=1.5)
plt.title('Distribuição de Medalhas por Esporte e Sexo')
plt.show()
```


Neste gráfico, observamos que todos os esportes analisados apresentam uma predominância masculina na distribuição de medalhas. Isso pode refletir várias dinâmicas e fatores subjacentes no mundo do esporte, que influenciam a participação e o sucesso entre os gêneros.

A predominância masculina em todas as modalidades pode indicar diferenças estruturais e culturais que impactam a representação e o desempenho das mulheres no esporte. A análise da distribuição de medalhas por sexo também pode fornecer informações sobre a necessidade de políticas mais inclusivas e iniciativas que promovam a igualdade de oportunidades. Avaliar as estratégias e ações que têm sido bem-sucedidas na promoção de esportes femininos pode ajudar a identificar melhores práticas e áreas para desenvolvimento futuro.

#### 6 - Medalhas por país 

```python
# Agrupando medalhas por país
contador_medalhas = df.groupby('Team')['Medal'].count().reset_index()

# Ordenar os dados para mostrar os países com mais medalhas no topo
contador_medalhas = contador_medalhas.sort_values(by='Medal', ascending=False)

# Fazer um recorte para os 20 maiores
recorte_20= contador_medalhas.head(20)

# Criar o gráfico interativo com Plotly
fig = px.bar(recorte_20, 
             x='Team', 
             y='Medal', 
             title='Top 20 Países com Mais Medalhas',
             labels={'Team': 'País', 'Medal': 'Número de Medalhas'},
             color='Medal',
             color_continuous_scale='Viridis')

fig.show()
```

Neste gráfico, observamos que os países com o maior número de medalhas são Estados Unidos (United States), União Soviética (Soviet Union), Alemanha (Germany), Reino Unido (Great Britain), França (France), Itália (Italy), Austrália (Australia) e Hungria (Hungary). Esses países têm desempenhado um papel dominante no cenário esportivo internacional, consolidando-se como potências ao longo dos anos.

A liderança dos Estados Unidos pode ser atribuída a uma combinação de fatores, incluindo grandes investimentos em programas esportivos, uma vasta infraestrutura e uma cultura que promove fortemente o esporte em diversos níveis. O Reino Unido e a União Soviética também têm tradições esportivas robustas e históricos de sucesso em competições internacionais, refletindo a continuidade do investimento em treinamento e desenvolvimento de atletas.

Essa análise de medalhas por países reflete não apenas a habilidade dos atletas, mas também o investimento, infraestrutura e programas de treinamento de elite que sustentam o sucesso dessas nações. Além disso, cada um desses países tem uma cultura esportiva enraizada que incentiva a prática esportiva desde tenra idade, o que é crucial para a formação de atletas de alto desempenho.

#### 7 - Distribuição de Medalhas por Idade, Esporte e País

```python
medal_age_sport = df.dropna(subset=['Medal']).groupby(['Age', 'Sport', 'Team']).size().reset_index(name='Medal Count')

# Criar o gráfico de dispersão
fig = px.scatter(medal_age_sport, 
                 x='Age', 
                 y='Medal Count', 
                 color='Team', 
                 size='Medal Count',
                 hover_data=['Sport'],
                 title='Distribuição de Medalhas por Idade, Esporte e País',
                 labels={'Age': 'Idade', 'Medal Count': 'Número de Medalhas', 'Team': 'País'},
                 size_max=15)
fig.show()
```

Neste gráfico, observamos a relação entre a distribuição de medalhas considerando três variáveis principais: a idade dos atletas, o esporte que praticam e o país que representam.

Idade: Em muitos esportes, notamos uma faixa etária predominante entre os medalhistas, geralmente entre 20 e 30 anos, que é considerada o auge do desempenho físico. No entanto, em alguns esportes, como ginástica e natação, atletas mais jovens frequentemente se destacam, enquanto em modalidades de resistência ou precisão, como tiro e esgrima, atletas mais experientes podem prevalecer. Essa variação mostra como o tempo de maturação e o pico de desempenho podem diferir de acordo com as exigências específicas de cada esporte.

Esporte: Os esportes mais tradicionais, como atletismo, ginástica e natação, tendem a ter maior diversidade etária e uma alta participação entre países de ponta, como Estados Unidos, União Soviética (ou seus sucessores) e Alemanha. Modalidades como esgrima, tiro e remo também exibem uma presença significativa entre medalhistas, embora com um perfil etário mais variado, destacando a importância da técnica e da estratégia.

País: Os principais países no quadro de medalhas, como Estados Unidos, União Soviética (URS), Alemanha, Reino Unido, França e Austrália, mantêm uma presença constante em diversas modalidades esportivas. A diversidade etária entre os medalhistas desses países reflete o desenvolvimento estruturado de atletas ao longo de diferentes fases de suas vidas, permitindo que tanto jovens  quanto atletas veteranos tenham oportunidades de se destacar. Países com tradições específicas em determinados esportes (como natação para Austrália e ginástica para Estados Unidos e URSS) demonstram maior concentração de medalhas em faixas etárias alinhadas às exigências físicas e técnicas dessas modalidades.

As análises realizadas fornecem uma visão abrangente sobre o cenário esportivo global, destacando padrões significativos e áreas para desenvolvimento futuro.

Em resumo, as análises mostram uma complexa rede de fatores que influenciam o sucesso esportivo. Compreender esses padrões pode ajudar a formular políticas e estratégias para promover a inclusão, apoiar o desenvolvimento de atletas em diferentes fases de suas carreiras e aumentar a competitividade global em uma variedade de esportes. Continuar a explorar e a abordar essas variáveis é crucial para garantir um ambiente esportivo mais equitativo e eficiente.
